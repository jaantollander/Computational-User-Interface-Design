```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
from A_5_1_one_euro_filter import OneEuroFilter
```

```python
df = pandas.read_csv("A_5_1_noise.csv")
t, x = df.values[0:200, 0], df.values[0:200, 1]
```

```python
plt.figure()
x2 = np.zeros_like(x)
x2[0] = x[0]
one_euro_filter = OneEuroFilter(
    t[0], x[0],
    min_cutoff=0.004,
    beta=0.005
)
for i in range(1, len(t)):
    x2[i] = one_euro_filter(t[i], x[i])
plt.plot(t, x)
plt.plot(t, x2)
plt.savefig("filtered_signal.png", dpi=300)
```
